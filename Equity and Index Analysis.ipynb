# 2. Data Preprocessing Process


## 2.1 Index Analysis: Methodology, Outputs, and Interpretation


### 2.1.1 Objectives


- Ingest heterogeneous index workbooks and produce a single aligned panel.
- Denoise and transform prices into stationary series suitable for analysis.
- Engineer technical features and anomaly flags for monitoring and modelling.
- Extract regime signals and cross-sectional structure (ranks, PCA).
- Link the preprocessed data to macro and cross-asset context.



### 2.1.2 Data Inputs & Assumptions


- Input directory: `FULL Nexus Data/*.xlsx` spanning global equities, funds, credit, rates, commodities, and macro factors exported from the Nexus data pack.
- Workbooks are ingested with `load_workbook_to_frame` plus `canonicalize_index_frame` to detect the date column and coerce numeric series.
- A primary value column is auto-selected from common aliases (e.g., adjusted close, close, px_last).
- Data are aligned to business-day frequency; limited forward-fill/back-fill covers brief gaps while preserving sparsity flags.
- A `source_catalog` DataFrame stores filename-to-category metadata for downstream summaries and filtering.


### 2.1.3 Pipeline Map (What each stage does)


- Stage 0 — Helpers: Reusable utilities for filling, filtering, features, regimes.
- Stage 1 — Cleaning & Alignment: Canonicalise frames, choose primary column, align to business days, and summarise coverage.
- Stages 2–3 — Noise Reduction & Stationarity: Moving-average and Fourier low-pass smoothing; transform to log-returns or differences; scale by rolling volatility.
- Stages 4–5 — Features & Anomalies: Rolling moments, RSI/MACD/Bollinger Bands, higher moments, winsorised returns, z-score anomaly flags.
- Stages 6–8 — Selection, Regimes, Cross-Section: PCA exposures, optional mutual-information ranking vs a target (e.g., SPX), regime detection, dynamic correlations, cross-sectional ranks.
- Stages 9–10 — Augmentation & Bundle: Optional block bootstrap for stress testing; assemble everything into `nexus_index_bundle` for easy reuse.
- Market Dynamics — Interpretation: Macro linkage, lead–lag, regime diagnostics, and behavioural proxies.



### 2.1.4 Key Artefacts (Where to look)


- `aligned_panel` (Stage 1): Unified DataFrame of primary value series indexed by date.
- `stage1_summary` (Stage 1): Per-series start/end, observation count, and data `coverage`.
- `stage2_smoothed` (Stage 2): Smoothed versions of the aligned panel (moving average and Fourier low-pass).
- `stationary_bundle` (Stage 3): `{ 'returns', 'diff', 'rolling_vol_20', 'vol_scaled' }`.
- `stage4_feature_panel` (Stage 4): Technical indicators per series (may be a MultiIndex on columns).
- `anomaly_flags`, `higher_moments`, `winsorized_returns` (Stage 5).
- `pca_features`, `mutual_information_rank` (Stage 6).
- `regimes`, `regime_scaled_returns`, `dynamic_corr_to_spx` (Stage 7).
- `cross_sectional_rank` (Stage 8).
- `bootstrap_samples`, `bootstrap_summary` (Stage 9).
- `nexus_index_bundle` (Stage 10): A dictionary exposing all of the above for downstream use.



### 2.1.5 How To Read The Results


- Coverage: In `stage1_summary`, prefer series with higher `coverage` for robust analysis; flag sparse series for manual review.
- Stationarity: Confirm `returns` and/or `diff` are mean-reverting enough for your modelling horizon; check `rolling_vol_20` for stability.
- Features: Inspect RSI, MACD, Bollinger Bands alongside price to validate signal behaviour (e.g., RSI extremes).
- Anomalies: Large z-scores highlight jumps/spikes; compare outcomes in `winsorized_returns` to control for outliers.
- PCA/Selection: Use `pca_features` to understand common drivers; `mutual_information_rank` can highlight series most informative to a chosen target.
- Regimes: Validate `regimes` (bull/bear/stagnant) against known episodes; use `regime_scaled_returns` for regime-aware normalisation.
- Cross-Asset: Check `dynamic_corr_to_spx` and `lead_lag_table` for propagation chains.
- Macro: Use `macro_beta_table` and `macro_correlations` (built in Market Dynamics) to sanity-check factor directionality and strength.



### 2.1.6 Limitations & Notes


- Smoothing and differencing choices can affect signal timing; verify against your use case.
- Mutual information ranking falls back to correlation if scikit-learn is unavailable.
- Regime detection uses simple moving-average comparisons; consider more advanced state models if needed.
- Bootstrap is illustrative and path-length matched; adapt block sizes to your series memory.


```python
from pathlib import Path
from openpyxl import load_workbook
import math
import numpy as np
import pandas as pd
from pandas._libs import lib as _pandas_lib
from pandas._libs import missing as _pandas_missing
try:
    from pandas._libs.tslibs.nattype import NaTType as _PANDAS_NATTYPE
except ImportError:
    _PANDAS_NATTYPE = type(pd.NaT)

# Patch pandas object conversion bug in this environment.

def _ensure_pandas_object_conversion():
    test_array = np.array([None], dtype=object)
    try:
        _pandas_lib.maybe_convert_objects(test_array)
    except TypeError as exc:
        if "Cannot convert numpy.ndarray to numpy.ndarray" in str(exc):
            def _safe_maybe_convert_objects(
                values,
                convert_dates=True,
                convert_numeric=True,
                convert_timedeltas=True,
                dtype=None,
                convert_non_numeric=False,
                **kwargs,
            ):
                if dtype is not None:
                    try:
                        return np.asarray(values, dtype=dtype)
                    except (TypeError, ValueError):
                        pass
                return np.asarray(values, dtype=object)
            _pandas_lib.maybe_convert_objects = _safe_maybe_convert_objects

_ensure_pandas_object_conversion()

# Harden pandas missing-value detection for this environment.
def _set_safe_isnaobj():
    def _value_is_missing(value, inf_as_na=False):
        if value is None:
            return True
        if value is pd.NA:
            return True
        if value is pd.NaT:
            return True
        if isinstance(value, _PANDAS_NATTYPE):
            return True
        if value is np.ma.masked:
            return True
        if isinstance(value, (float, np.floating)):
            if math.isnan(value):
                return True
            if inf_as_na and math.isinf(value):
                return True
        if isinstance(value, np.generic):
            if isinstance(value, (np.datetime64, np.timedelta64)):
                return np.isnat(value)
            try:
                if np.isnan(value):
                    return True
            except (TypeError, ValueError):
                pass
            if inf_as_na:
                try:
                    if np.isinf(value):
                        return True
                except (TypeError, ValueError):
                    pass
        if isinstance(value, (np.datetime64, np.timedelta64)):
            return np.isnat(value)
        if hasattr(value, "is_nan"):
            try:
                return bool(value.is_nan())
            except Exception:
                pass
        return False

    def _safe_isnaobj(values, inf_as_na=False):
        arr = np.asarray(values, dtype=object)
        if arr.size == 0:
            return np.zeros(arr.shape, dtype=bool)
        flat = [_value_is_missing(item, inf_as_na=inf_as_na) for item in arr.ravel()]
        return np.array(flat, dtype=bool).reshape(arr.shape)

    _pandas_missing.isnaobj = _safe_isnaobj

_set_safe_isnaobj()


from collections import defaultdict
import re
from IPython.display import Markdown, display

data_folder = Path('FULL Nexus Data')
xlsx_files = sorted(data_folder.glob('*.xlsx'))

if not xlsx_files:
    raise FileNotFoundError(f'No .xlsx files found in {data_folder.resolve()}')

def sanitize_cell(value):
    if isinstance(value, (list, tuple)):
        for item in value:
            cleaned = sanitize_cell(item)
            if cleaned is not None:
                return cleaned
        return None
    if isinstance(value, np.ndarray):
        if value.size == 0:
            return None
        return sanitize_cell(value.flat[0])
    if isinstance(value, np.generic):
        return value.item()
    return value

def normalize_value(value):
    cleaned = sanitize_cell(value)
    if isinstance(cleaned, (np.ndarray, list, tuple, dict, set)):
        cleaned = str(cleaned)
    if isinstance(cleaned, str):
        stripped = cleaned.strip()
        if stripped in {"", "#N/A", "#N/A N/A"}:
            return None
        return stripped
    return cleaned

def is_missing(value):
    if value is None:
        return True
    if value is pd.NA or value is pd.NaT:
        return True
    if isinstance(value, _PANDAS_NATTYPE):
        return True
    if isinstance(value, (float, np.floating)):
        if math.isnan(value):
            return True
    if isinstance(value, np.generic):
        if isinstance(value, (np.datetime64, np.timedelta64)):
            return np.isnat(value)
        try:
            if np.isnan(value):
                return True
        except (TypeError, ValueError):
            pass
    if isinstance(value, (np.datetime64, np.timedelta64)):
        return np.isnat(value)
    if hasattr(value, "is_nan"):
        try:
            return bool(value.is_nan())
        except Exception:
            pass
    return False


# Safe fallback for DataFrame.dropna when pandas internals misbehave.
def _dataframe_to_object_matrix(df: pd.DataFrame) -> np.ndarray:
    n_rows = len(df.index)
    n_cols = len(df.columns)
    matrix = np.empty((n_rows, n_cols), dtype=object)
    for block in df._mgr.blocks:
        values = np.asarray(block.values, dtype=object)
        locs = block.mgr_locs.as_array
        for offset, col_pos in enumerate(locs):
            matrix[:, col_pos] = values[offset]
    return matrix

if getattr(pd.DataFrame.dropna, "__name__", "") == "_safe_dataframe_dropna":
    _ORIGINAL_DF_DROPNA = getattr(pd.DataFrame.dropna, "_safe_dropna_original", pd.DataFrame.dropna)
else:
    _ORIGINAL_DF_DROPNA = pd.DataFrame.dropna

def _safe_dataframe_dropna(self, *args, **kwargs):
    try:
        return _ORIGINAL_DF_DROPNA(self, *args, **kwargs)
    except (TypeError, SystemError, RuntimeError):
        axis = kwargs.get("axis", 0)
        if axis in ("index", None):
            axis = 0
        if axis not in (0,):
            raise NotImplementedError("safe dropna only supports axis=0 in this environment")
        how = kwargs.get("how", "any")
        thresh = kwargs.get("thresh")
        subset = kwargs.get("subset")
        inplace = kwargs.get("inplace", False)
        columns = list(self.columns)
        subset_cols = columns if subset is None else ([subset] if isinstance(subset, str) else list(subset))
        matrix = _dataframe_to_object_matrix(self)
        subset_positions = [columns.index(col) for col in subset_cols]
        keep_positions = []
        total_rows = len(self.index)
        for row_idx in range(total_rows):
            values = [matrix[row_idx, col_pos] for col_pos in subset_positions]
            missing_flags = [is_missing(val) for val in values]
            if thresh is not None:
                non_missing = len(values) - sum(missing_flags)
                if non_missing >= thresh:
                    keep_positions.append(row_idx)
                continue
            if how == "all":
                if not all(missing_flags):
                    keep_positions.append(row_idx)
            else:
                if not any(missing_flags):
                    keep_positions.append(row_idx)
        def _coerce_scalar(value):
            if isinstance(value, (list, tuple)):
                for item in value:
                    coerced = _coerce_scalar(item)
                    if coerced is not None:
                        return coerced
                return None
            if isinstance(value, np.ndarray):
                if value.size == 0:
                    return None
                return _coerce_scalar(value.flat[0])
            if isinstance(value, np.generic):
                return value.item()
            return value

        if keep_positions:
            new_index = [self.index[pos] for pos in keep_positions]
            reduced_rows = [
                [_coerce_scalar(matrix[row_idx, col_idx]) for col_idx in range(len(columns))]
                for row_idx in keep_positions
            ]
            result = pd.DataFrame(reduced_rows, columns=columns, index=new_index)
        else:
            result = pd.DataFrame(columns=columns, index=self.index[:0])
        if inplace:
            self._mgr = result._mgr
            self._clear_item_cache()
            return None
        return result

_safe_dataframe_dropna._safe_dropna_original = _ORIGINAL_DF_DROPNA
pd.DataFrame.dropna = _safe_dataframe_dropna

def load_workbook_to_frame(path: Path) -> pd.DataFrame:
    wb = load_workbook(path, read_only=True, data_only=True)
    ws = wb.active
    raw_rows: list[list[object]] = []
    max_width = 0
    for row in ws.iter_rows(values_only=True):
        processed = [normalize_value(cell) for cell in row]
        if processed:
            max_width = max(max_width, len(processed))
        raw_rows.append(processed)
    wb.close()
    if not raw_rows:
        return pd.DataFrame()

    padded = [row + [None] * (max_width - len(row)) for row in raw_rows]

    filtered_rows = [row for row in padded if any(not is_missing(val) for val in row)]
    if not filtered_rows:
        return pd.DataFrame()

    keep_indices = [
        idx for idx in range(max_width)
        if any(not is_missing(row[idx]) for row in filtered_rows)
    ]
    if not keep_indices:
        return pd.DataFrame()

    trimmed = [[row[idx] for idx in keep_indices] for row in filtered_rows]
    columns = [f"col_{i}" for i in range(len(keep_indices))]
    frame = pd.DataFrame(trimmed, columns=columns)
    frame = frame.astype(object)
    return frame

macro_tokens = {
    'gdp', 'cpi', 'ppi', 'pmi', 'm2', 'industrial', 'sentiment', 'confidence',
    'surprise', 'retail', 'goods', 'housing', 'nhspatot', 'ip', 'nfp',
    'unemployment', 'rstax', 'dgnochng', 'conssent', 'concc', 'usurtot', 'cesi',
    'manufacturing', 'authorization', 'authorized', 'sales', 'production',
}

def infer_category(name: str) -> str:
    full_lower = name.lower()
    left, right = name.split(' - ', 1)
    tail_lower = right.lower()
    tail_tokens = set(re.findall(r'[a-z0-9]+', tail_lower))
    full_tokens = set(re.findall(r'[a-z0-9]+', full_lower))

    if any(token in full_tokens for token in ['bitcoin', 'ethereum', 'crypto', 'bgci', 'xbt', 'xet']):
        return 'Digital Assets'
    if 'cds' in tail_tokens:
        return 'Credit Derivatives'
    if {'fund', 'etf', 'shares', 'multi', 'hedge'} & full_tokens:
        return 'Funds & ETFs'
    if {'futur', 'future', 'fwd'} & full_tokens:
        return 'Futures & Forwards'
    if 'comdty' in tail_tokens:
        return 'Commodities'
    if 'curncy' in tail_tokens:
        return 'FX & Rates'
    if 'corp' in tail_tokens and 'conv' in tail_tokens:
        return 'Convertible Credit'
    if 'corp' in tail_tokens:
        return 'Corporate Credit'
    if 'govt' in tail_tokens:
        return 'Government Bonds'
    if 'muni' in tail_tokens:
        return 'Municipal Bonds'
    if any(re.search(r' [pc]\d{3,}', part.lower()) for part in (left, right)):
        return 'Options & Derivatives'
    if 'opt' in full_tokens or 'option' in full_tokens:
        return 'Options & Derivatives'
    if 'equity' in tail_tokens:
        return 'Equities'
    if 'index' in tail_tokens:
        if macro_tokens & (tail_tokens | full_tokens):
            return 'Macro Indicators'
        if {'vix', 'v2x', 'move'} & full_tokens:
            return 'Volatility Indices'
        return 'Market Indices'
    if {'vix', 'v2x', 'move'} & full_tokens:
        return 'Volatility Indices'
    if {'sofr', 'estr', 'shibor', 'libor'} & full_tokens:
        return 'FX & Rates'
    if {'usd', 'eur', 'aud'} & tail_tokens:
        return 'FX & Rates'
    return 'Other'


def _build_summary_fallback(rows):
    groups = defaultdict(list)
    for entry in rows:
        groups[entry['category']].append(entry['series'])
    ordered = sorted(groups.items(), key=lambda item: len(item[1]), reverse=True)
    records = []
    for category, series_names in ordered:
        sample = ", ".join(series_names[:3])
        if not sample:
            sample = '—'
        records.append((category, {'total': len(series_names), 'sample': sample}))
    return records



nexus_raw_data: dict[str, pd.DataFrame] = {}
catalog_rows: list[dict[str, str]] = []
for path_obj in xlsx_files:
    series_name = path_obj.stem
    raw_frame = load_workbook_to_frame(path_obj)
    nexus_raw_data[series_name] = raw_frame
    catalog_rows.append({
        'series': series_name,
        'filename': path_obj.name,
        'category': infer_category(series_name),
    })

source_catalog = pd.DataFrame(catalog_rows)

summary_records = None
try:
    summary = (
        source_catalog.groupby('category')['series']
        .agg(total='count', sample=lambda s: ', '.join(s.head(3)))
        .sort_values('total', ascending=False)
    )
except TypeError as exc:
    if 'Cannot convert numpy.ndarray to numpy.ndarray' not in str(exc):
        raise
    summary = None
    summary_records = _build_summary_fallback(catalog_rows)
summary_lines = [
    '### FULL Nexus Data Inventory',
    f'Loaded {len(source_catalog)} spreadsheets from `{data_folder}`.'
]

summary_iter = summary.iterrows() if summary is not None else iter(summary_records)
for category, row in summary_iter:
    sample = row['sample'] or ''
    if row['total'] > 3 and sample:
        sample += ', …'
    if not sample:
        sample = '—'
    summary_lines.append(f"- **{category}** ({int(row['total'])} series) — {sample}")

display(Markdown('\n'.join(summary_lines)))
print('First few entries:', list(nexus_raw_data.keys())[:5])
```

## 2.2 Stage 0 - Helpers

Purpose: provide reusable utilities for resampling, smoothing, transformations, feature construction, anomaly flags, regime labelling, dimensionality reduction, and basic augmentation. Defaults are conservative and documented below.

- Resampling & cleaning
  - `forward_backward_fill(df, columns=None)`: ffill then bfill selected columns; avoids in‑place mutation.
  - `align_to_frequency(df, date_col, freq, method="ffill")`: resample to e.g. business day (`"B"`), using `ffill`/`bfill`/interpolation.
  - `canonicalize_index_frame(df)`: standardize column names, detect a date column, keep numeric value columns, drop duplicates.
  - `pick_primary_column(columns)`: choose the main value column from common aliases.

- Core transforms
  - `compute_log_returns(series)`: `ln(P_t/P_{t-1})`, treating non‑positive values as missing.
  - `volatility_scale(returns, vol)`: divide by rolling volatility (ex‑post) to reduce heteroscedasticity.
  - `winsorize(series, q1=0.01, q2=0.99)`: clip tails to robustify against outliers.

- Technical features
  - `rolling_window_features(series, window=20)`: rolling mean/std.
  - `compute_rsi(prices, window=14)`: EWM RSI(14).
  - `compute_macd(prices, 12, 26, 9)`: EMA fast/slow, signal, histogram.
  - `compute_bbands(prices, window=20, n_std=2)`: mid/upper/lower bands.
  - `rolling_higher_moments(returns, window=60)`: rolling skew/kurtosis.

- Regimes & dependence
  - `detect_market_regimes(series, fast=20, slow=100)`: MA crossover → bull/bear/stagnant.
  - `regime_specific_normalize(series, regimes)`: z‑score within contiguous regime segments.
  - `dynamic_correlation(a, b, window=60)`: rolling Pearson correlation.

- Cross‑section & dimensionality reduction
  - `principal_component_features(features, n_components≤5)`: eigen‑decompose covariance and project onto top PCs.
  - `mutual_information_ranking(features, target)`: MI vs target (fallback to |corr|).
  - `cross_sectional_rank_normalize(df, date_col, value_col)`: per‑date ranks mapped to (0,1).

- Augmentation
  - `block_bootstrap(series, block_size=5, n_samples=100)`: sample fixed‑size blocks to preserve autocorrelation.

Notes:
- Most functions preserve the original index and return DataFrames aligned to inputs.
- Rolling windows use `min_periods=window` to avoid biased early values; expect initial NaNs.
- Replace or tune defaults (windows, cutoffs) to match your horizon and liquidity.


```python
import numpy as np
from typing import Dict, List, Optional, Sequence


#
# - Fill NaNs forward then backward for specified columns.
# - Returns a copy to avoid mutating input frame.
def forward_backward_fill(frame: pd.DataFrame, columns: Optional[Sequence[str]] = None) -> pd.DataFrame:
    columns = list(columns) if columns is not None else list(frame.columns)
    out = frame.copy()
    out[columns] = out[columns].ffill().bfill()
    return out


#
# - Resample by `freq` using ffill/bfill or interpolation.
# - Keeps `date_col` as a column after reset_index().
def align_to_frequency(df: pd.DataFrame, date_col: str, freq: str, method: str = "ffill") -> pd.DataFrame:
    out = df.set_index(date_col).sort_index()
    if method == "ffill":
        out = out.resample(freq).ffill()
    elif method == "bfill":
        out = out.resample(freq).bfill()
    else:
        out = out.resample(freq).interpolate(method=method)
    return out.reset_index()


#
# - Simple rolling mean denoiser with min_periods=1.
def moving_average_filter(series: pd.Series, window: int = 5) -> pd.Series:
    return series.rolling(window=window, min_periods=1, center=False).mean()


#
# - Zero high-frequency FFT coefficients beyond cutoff_ratio.
# - Inverse transform and preserve original index length.
def fourier_lowpass_filter(series: pd.Series, cutoff_ratio: float = 0.1) -> pd.Series:
    values = series.to_numpy()
    n = values.shape[0]
    if n == 0:
        return series.copy()
    fft = np.fft.rfft(values)
    cutoff = int(len(fft) * cutoff_ratio)
    filtered_fft = np.zeros_like(fft)
    filtered_fft[:max(cutoff, 1)] = fft[:max(cutoff, 1)]
    filtered = np.fft.irfft(filtered_fft, n=n)
    return pd.Series(filtered, index=series.index)


#
# - Standardize to mean 0, std 1; guard for std==0.
def zscore_normalize(series: pd.Series) -> pd.Series:
    mean = series.mean()
    std = series.std()
    if std == 0:
        return series - mean
    return (series - mean) / std


#
# - Log difference of levels; first observation is NaN.
# - Ensure inputs are positive/non-zero upstream to avoid infs.
def compute_log_returns(series: pd.Series) -> pd.Series:
    return np.log(series).diff()


#
# - Scale returns by provided volatility; zero vol -> NaN.
def volatility_scale(returns: pd.Series, vol: pd.Series) -> pd.Series:
    return returns / vol.replace(0.0, np.nan)


#
# - Clip series at given lower/upper quantiles to reduce outliers.
def winsorize(series: pd.Series, lower_q: float = 0.01, upper_q: float = 0.99) -> pd.Series:
    lo, hi = series.quantile([lower_q, upper_q])
    return series.clip(lower=lo, upper=hi)


#
# - Return DataFrame of rolling mean and std.
def rolling_window_features(series: pd.Series, window: int = 20) -> pd.DataFrame:
    roll_mean = series.rolling(window=window, min_periods=window).mean()
    roll_std = series.rolling(window=window, min_periods=window).std()
    return pd.DataFrame({"rolling_mean": roll_mean, "rolling_std": roll_std})


#
# - EWM-based RSI with default 14 periods (0–100).
def compute_rsi(prices: pd.Series, window: int = 14) -> pd.Series:
    delta = prices.diff()
    gain = delta.clip(lower=0).ewm(alpha=1 / window, adjust=False, min_periods=window).mean()
    loss = -delta.clip(upper=0).ewm(alpha=1 / window, adjust=False, min_periods=window).mean()
    rs = gain / loss.replace(0, np.nan)
    rsi = 100 - (100 / (1 + rs))
    return rsi


#
# - EMA(12/26) MACD, 9-period signal, and histogram.
def compute_macd(prices: pd.Series, span_fast: int = 12, span_slow: int = 26, span_signal: int = 9) -> pd.DataFrame:
    ema_fast = prices.ewm(span=span_fast, adjust=False).mean()
    ema_slow = prices.ewm(span=span_slow, adjust=False).mean()
    macd_line = ema_fast - ema_slow
    signal = macd_line.ewm(span=span_signal, adjust=False).mean()
    hist = macd_line - signal
    return pd.DataFrame({"macd": macd_line, "macd_signal": signal, "macd_hist": hist})


#
# - Rolling mean ± n_std*std; returns mid/upper/lower bands.
def compute_bbands(prices: pd.Series, window: int = 20, n_std: float = 2.0) -> pd.DataFrame:
    ma = prices.rolling(window=window, min_periods=window).mean()
    sd = prices.rolling(window=window, min_periods=window).std()
    upper = ma + n_std * sd
    lower = ma - n_std * sd
    return pd.DataFrame({"bb_mid": ma, "bb_upper": upper, "bb_lower": lower})


#
# - Rolling skewness and kurtosis over window (default 60).
def rolling_higher_moments(returns: pd.Series, window: int = 60) -> pd.DataFrame:
    skew = returns.rolling(window=window, min_periods=window).skew()
    kurt = returns.rolling(window=window, min_periods=window).kurt()
    return pd.DataFrame({"rolling_skew": skew, "rolling_kurt": kurt})


#
# - MA crossover labels: bull (fast>slow), bear (fast<slow), else stagnant.
def detect_market_regimes(series: pd.Series, fast_window: int = 20, slow_window: int = 100) -> pd.Series:
    fast_ma = series.rolling(window=fast_window, min_periods=1).mean()
    slow_ma = series.rolling(window=slow_window, min_periods=1).mean()
    regime = pd.Series("stagnant", index=series.index)
    regime = regime.mask(fast_ma > slow_ma, "bull")
    regime = regime.mask(fast_ma < slow_ma, "bear")
    return regime


#
# - Z-score within contiguous regime segments; preserves index.
def regime_specific_normalize(series: pd.Series, regimes: pd.Series) -> pd.Series:
    series = series.copy()
    regimes = regimes.reindex(series.index)
    out = series.copy()
    for regime, idx in regimes.groupby(regimes).groups.items():
        labels = list(idx)
        normalized = zscore_normalize(series.loc[labels])
        out.loc[labels] = normalized
    return out


#
# - Binary flag where |z| exceeds threshold (default 3.0).
def detect_anomalies_zscore(series: pd.Series, z_thresh: float = 3.0) -> pd.Series:
    z = zscore_normalize(series)
    return (np.abs(z) > z_thresh).astype(int)


#
# - Rolling correlation with min_periods=window.
def dynamic_correlation(series_a: pd.Series, series_b: pd.Series, window: int = 60) -> pd.Series:
    return series_a.rolling(window=window, min_periods=window).corr(series_b)


#
# - Center features, eigendecompose covariance, return top PCs.
def principal_component_features(features: pd.DataFrame, n_components: int = 3) -> pd.DataFrame:
    centered = features - features.mean()
    cov = np.cov(centered.fillna(0.0).to_numpy().T)
    eigvals, eigvecs = np.linalg.eigh(cov)
    order = np.argsort(eigvals)[::-1][:n_components]
    pcs = centered.to_numpy() @ eigvecs[:, order]
    columns = [f"pc{i+1}" for i in range(len(order))]
    return pd.DataFrame(pcs, index=features.index, columns=columns)


#
# - Rank features by MI vs target (fallback: |corr|).
def mutual_information_ranking(features: pd.DataFrame, target: pd.Series) -> pd.Series:
    try:
        from sklearn.feature_selection import mutual_info_regression  # type: ignore
    except Exception:
        corr = features.corrwith(target)
        return corr.abs().sort_values(ascending=False)
    mi = mutual_info_regression(features.fillna(0.0), target.fillna(0.0))
    return pd.Series(mi, index=features.columns).sort_values(ascending=False)


#
# - Per-date ranks scaled to (0,1) via average method.
def cross_sectional_rank_normalize(df: pd.DataFrame, date_col: str, value_col: str) -> pd.Series:
    ranks = df.groupby(date_col)[value_col].rank(method="average")
    group_sizes = df.groupby(date_col)[value_col].transform("count")
    return (ranks - 0.5) / group_sizes


#
# - Sample fixed-size blocks to preserve autocorrelation structure.
def block_bootstrap(series: pd.Series, block_size: int = 5, n_samples: int = 100, random_state: Optional[int] = None) -> np.ndarray:
    if block_size <= 0:
        raise ValueError('block_size must be positive')
    rng = np.random.default_rng(random_state)
    values = series.dropna().to_numpy()
    if len(values) == 0:
        return np.empty((n_samples, 0))
    blocks = [values[i:i + block_size] for i in range(0, len(values), block_size)]
    if not blocks:
        return np.empty((n_samples, 0))
    n_blocks_needed = max(int(np.ceil(len(values) / block_size)), 1)
    bootstrapped = np.empty((n_samples, len(values)))
    for sample_idx in range(n_samples):
        picks = rng.integers(0, len(blocks), size=n_blocks_needed)
        stitched = np.concatenate([blocks[idx] for idx in picks])
        while stitched.shape[0] < len(values):
            extra_idx = int(rng.integers(0, len(blocks)))
            stitched = np.concatenate([stitched, blocks[extra_idx]])
        bootstrapped[sample_idx, :] = stitched[:len(values)]
    return bootstrapped

```

## 2.3 Stage 1 - Data Cleaning and Alignment


Canonicalize raw workbooks, resample to business days, pick a primary value column, and assemble an aligned panel with a coverage audit.

- Input handling
  - Read `INDEX/*.xlsx` into DataFrames; `canonicalize_index_frame` parses dates, keeps numeric columns, drops duplicates.
  - `pick_primary_column` selects the main value series per file (close/adj close/price aliases).

- Alignment
  - Resample to business‑day frequency with forward/backward fill as needed via `align_to_frequency`.
  - Build `aligned_panel`: one column per series, indexed by the union of business days across inputs.

- Summary
  - `stage1_summary`: per‑series `[start, end, observations]`.
  - `coverage`: share of non‑missing days in `aligned_panel`; use to flag sparse series.

- Caveats
  - Sparse or irregular sources can reduce effective overlap post‑alignment; downstream steps drop all‑NaN rows as needed.
  - Forward fills are pragmatic for monthly/quarterly macro series; confirm economic suitability for each series.


```python
def canonicalize_index_frame(df: pd.DataFrame) -> pd.DataFrame:
    frame = df.copy()
    frame = frame.dropna(how="all")
    frame.columns = [
        (str(col).strip() if not str(col).startswith("Unnamed") else f"col_{i}")
        for i, col in enumerate(frame.columns)
    ]
    if frame.empty:
        return frame
    first_col = frame.columns[0]
    frame = frame.rename(columns={first_col: "date"})
    if isinstance(frame["date"].iloc[0], str) and "date" in frame["date"].iloc[0].lower():
        frame = frame.iloc[1:].reset_index(drop=True)
    frame["date"] = pd.to_datetime(frame["date"], errors="coerce")
    frame = frame.dropna(subset=["date"])
    numeric_cols = [col for col in frame.columns if col != "date"]

    def _ensure_scalar(value):
        if isinstance(value, (list, tuple)):
            return value[0] if len(value) == 1 else np.nan
        if isinstance(value, np.ndarray):
            return value.item() if value.size == 1 else np.nan
        return value

    for col in numeric_cols:
        frame[col] = frame[col].map(_ensure_scalar)
        frame[col] = pd.to_numeric(frame[col], errors="coerce")
    numeric_cols = [col for col in numeric_cols if not frame[col].isna().all()]
    frame = frame[["date"] + numeric_cols]
    if numeric_cols:
        frame = forward_backward_fill(frame.sort_values("date"), numeric_cols)
    frame = frame.drop_duplicates(subset=["date"]).reset_index(drop=True)
    return frame
```

## 2.4 Stages 2-3 - Noise Reduction and Stationarity


Apply smoothing, stationarity transforms, and volatility scaling to prepare features that behave well in simple models.

- Smoothing (Stage 2)
  - `moving_average_filter(window=5)`: reduces short‑term noise in levels.
  - `fourier_lowpass_filter(cutoff_ratio=0.1)`: trims high‑frequency components while preserving trend.
  - Output: `stage2_smoothed` with a MultiIndex columns `(filter, series)`.

- Stationarity (Stage 3)
  - `returns`: `ln(P_t/P_{t-1})` with non‑positive prices treated as missing.
  - `diff`: first differences for level series where log returns are inappropriate.
  - `rolling_vol_20`: 20‑day realized vol on returns.
  - `vol_scaled`: returns divided by `rolling_vol_20` to stabilize variance.
  - Output: `stationary_bundle = {returns, diff, rolling_vol_20, vol_scaled}`.

- Notes
  - Early windows contain NaNs by design; downstream code aligns and drops where necessary.
  - Choose `returns` vs `diff` based on data‑generating process (e.g., price vs rate).
  - Vol scaling is ex‑post; avoid using scaled values for causal inference without care.


```python
if aligned_panel.empty:
    raise ValueError("Stage 1 produced no aligned panel to process.")


#
# - Interpolate small gaps, apply Fourier low-pass, restore NaNs.
def _apply_lowpass(series: pd.Series) -> pd.Series:
    filled = series.interpolate(limit_direction="both")
    filtered = fourier_lowpass_filter(filled)
    filtered[series.isna()] = np.nan
    return filtered


moving_average_panel = aligned_panel.apply(moving_average_filter)
lowpass_panel = aligned_panel.apply(_apply_lowpass)

stage2_smoothed = pd.concat(
    {
        "moving_average": moving_average_panel,
        "fourier_lowpass": lowpass_panel,
    },
    axis=1,
)

returns = aligned_panel.apply(lambda s: compute_log_returns(s.replace(0.0, np.nan)))
returns = returns.replace([np.inf, -np.inf], np.nan)
diffs = aligned_panel.diff()
rolling_vol_20 = returns.rolling(window=20, min_periods=20).std()
vol_scaled_returns = pd.DataFrame({
    col: volatility_scale(returns[col], rolling_vol_20[col])
    for col in returns.columns
})

stationary_bundle = {
    "returns": returns,
    "diff": diffs,
    "rolling_vol_20": rolling_vol_20,
    "vol_scaled": vol_scaled_returns,
}

nexus_index_stages.update({
    "stage2": {"smoothed": stage2_smoothed},
    "stage3": stationary_bundle,
})

stage2_smoothed.tail()
```

## 2.5 Stages 4-5 - Feature Engineering and Anomalies


Construct technical features, robustify returns, and flag anomalies for monitoring and downstream modeling.

- Features (Stage 4)
  - Rolling stats: 20‑day mean/std on price.
  - RSI(14): EWM gains/losses, `RSI = 100 − 100/(1+RS)`.
  - MACD: EMA(12/26) with 9‑period signal and histogram.
  - Bollinger Bands: 20‑day mid ± 2σ.
  - Higher moments: 60‑day rolling skew/kurtosis on returns.
  - Output: `stage4_feature_panel` as a column‑MultiIndex `(series, feature)`.

- Robust returns & anomalies (Stage 5)
  - `winsorized_returns`: clip extreme scaled returns at 1st/99th percentiles.
  - `higher_moments`: per‑series skew/kurtosis panels (for diagnostics).
  - `anomaly_flags`: 1 where |z|>3 on winsorized returns; proxy for jumps/shocks.

- Usage
  - Combine features with regimes for context (e.g., RSI signals during bull vs bear).
  - Use anomaly statistics to calibrate stop rules or post‑event drift studies.


```python
winsorized_returns = vol_scaled_returns.apply(lambda s: winsorize(s))
higher_moments_panel = pd.concat(
    {
        col: rolling_higher_moments(returns[col]).rename(columns=lambda c: f"{c}")
        for col in returns.columns
    },
    axis=1,
)

technical_features = {}
for name in aligned_panel.columns:
    features = [
        rolling_window_features(aligned_panel[name]).add_prefix("roll_"),
        compute_rsi(aligned_panel[name]).to_frame("rsi14"),
        compute_macd(aligned_panel[name]).rename(columns=lambda c: f"macd_{c}"),
        compute_bbands(aligned_panel[name]).rename(columns=lambda c: f"bb_{c}"),
        rolling_higher_moments(returns[name]).rename(columns=lambda c: f"moment_{c}"),
    ]
    technical_features[name] = pd.concat(features, axis=1)

stage4_feature_panel = pd.concat(technical_features, axis=1)
anomaly_flags = winsorized_returns.apply(lambda s: detect_anomalies_zscore(s)).astype(int)

nexus_index_stages.update({
    "stage4": {"features": stage4_feature_panel},
    "stage5": {
        "winsorized_returns": winsorized_returns,
        "higher_moments": higher_moments_panel,
        "anomaly_flags": anomaly_flags,
    },
})

anomaly_flags.tail()
```

## 2.6 Stages 6-8 - Selection, Regimes, Cross-Section


Extract low‑dimensional structure, rank features by informativeness, detect regimes, assess time‑varying co‑movements, and produce cross‑sectional ranks.

- Dimensionality reduction (Stage 6)
  - PCA on filled returns (ffill/bfill) → `pc1..pcN` via covariance eigendecomposition.
  - `mutual_information_ranking`: rank features vs SPX (if present), fallback to |corr| when sklearn is unavailable.

- Regimes & dependence (Stage 7)
  - `regimes`: MA(20/100) crossover labels on price (bull/bear/stagnant).
  - `regime_scaled_returns`: z‑score within contiguous regime segments to compare signals across states.
  - `dynamic_corr_to_spx`: 60‑day rolling correlation of each series vs SPX returns.

- Cross‑section (Stage 8)
  - `cross_sectional_rank`: for each date, rank returns and scale to (0,1) for robust relative strength signals.

- Caveats
  - PCA loadings and MI ranks are sample‑dependent; validate across sub‑samples.
  - Rolling correlations are sensitive to window length and outliers.


```python
valid_returns = returns.dropna(how="all")
if not valid_returns.empty:
    filled_returns = valid_returns.fillna(method="ffill").fillna(method="bfill")
    n_components = min(5, max(1, filled_returns.shape[1]))
    pca_features = principal_component_features(filled_returns, n_components=n_components)
else:
    pca_features = pd.DataFrame()

spx_candidate = next((col for col in aligned_panel.columns if "spx" in col.lower() or "s&p" in col.lower()), None)
if spx_candidate and spx_candidate in returns.columns and returns.shape[1] > 1:
    features_for_mi = returns.drop(columns=[spx_candidate])
    mi_rank = mutual_information_ranking(features_for_mi, returns[spx_candidate])
else:
    mi_rank = None

regimes = pd.DataFrame({
    name: detect_market_regimes(aligned_panel[name])
    for name in aligned_panel.columns
})
regime_scaled_returns = pd.DataFrame({
    name: regime_specific_normalize(returns[name], regimes[name])
    for name in returns.columns
})

if spx_candidate and spx_candidate in returns.columns:
    dyn_corr = pd.DataFrame({
        name: dynamic_correlation(returns[name], returns[spx_candidate])
        for name in returns.columns
        if name != spx_candidate
    })
else:
    dyn_corr = pd.DataFrame()

stacked_returns = returns.stack().rename("return").reset_index()
stacked_returns.columns = ["date", "series", "return"]
stacked_returns["cs_rank"] = cross_sectional_rank_normalize(stacked_returns, "date", "return")
cross_sectional_panel = stacked_returns.pivot(index="date", columns="series", values="cs_rank")

nexus_index_stages.update({
    "stage6": {
        "pca": pca_features,
        "mutual_information_rank": mi_rank,
    },
    "stage7": {
        "regimes": regimes,
        "regime_scaled_returns": regime_scaled_returns,
        "dynamic_corr_to_spx": dyn_corr,
    },
    "stage8": {
        "cross_sectional_rank": cross_sectional_panel,
    },
})

pca_features.head()
```

## 2.7 Stages 9-10 - Augmentation and Pipeline Bundle


Augment data with simple block bootstrap for stress testing and consolidate all stage outputs for downstream use.

- Stage 9 — Augmentation
  - If SPX returns are available, generate `n_samples` block‑bootstrap paths with `block_size=5` to preserve serial dependence.

- Stage 10 — Bundle

```python
bootstrap_samples: Dict[str, np.ndarray] = {}
if spx_candidate and spx_candidate in returns.columns:
    bootstrap_samples[spx_candidate] = block_bootstrap(returns[spx_candidate], block_size=5, n_samples=200, random_state=42)
bootstrap_summary = {
    name: {
        "mean_mean": samples.mean(axis=1).mean(),
        "mean_std": samples.mean(axis=1).std(),
        "path_std_mean": samples.std(axis=1).mean(),
    }
    for name, samples in bootstrap_samples.items()
}

nexus_index_bundle = {
    "aligned_panel": aligned_panel,
    "smoothed": stage2_smoothed,
    "stationary": stationary_bundle,
    "features": stage4_feature_panel,
    "anomalies": anomaly_flags,
    "pca": pca_features,
    "regimes": regimes,
    "regime_scaled_returns": regime_scaled_returns,
    "dynamic_corr_to_spx": dyn_corr,
    "cross_sectional_rank": cross_sectional_panel,
    "bootstrap_samples": bootstrap_samples,
    "bootstrap_summary": bootstrap_summary,
}

nexus_index_stages.update({
    "stage9": {
        "bootstrap_samples": bootstrap_samples,
        "bootstrap_summary": bootstrap_summary,
    },
    "stage10": {
        "bundle": nexus_index_bundle,
    },
})

summary_table = pd.DataFrame({
    "stage": list(nexus_index_stages.keys()),
    "artifacts": [", ".join(sorted(map(str, section.keys()))) for section in nexus_index_stages.values()],
})
summary_table
```

# 3. Market Dynamics Analysis


Link the staged preprocessing outputs to macro drivers, cross-asset influence, regime diagnostics, and behavioural signals for portfolio context.

## Methodological Overview
This section describes how we interrogate market dynamics for the index using robust, reproducible diagnostics. The goal is to map short- and medium-horizon behaviour to macro exposures, cross-asset structure, regimes, seasonality, liquidity, and tails, and to summarize the current state with portfolio-relevant context.

### 1) Return & Volatility Structure
- Compute log returns `r_t = ln(P_t/P_{t-1})` and summarize distributional shape (mean, volatility, skew, kurtosis).
- Normality diagnostics: QQ-plot, Jarque–Bera; report outliers with robust winsorization bounds.
- Volatility: rolling realized volatility `rv_t = sqrt(annualization) * std(r_{t-w+1:t})` across windows (e.g., 21/63/252).
- Dependence: ACF/PACF of `r_t` and `r_t^2` with Ljung–Box to test serial correlation and volatility clustering.
- Optional: fit a simple GARCH(1,1) for conditional variance as a sanity check on clustering.

### 2) Trend vs. Mean Reversion
- Stationarity and persistence: ADF/KPSS on price and return; Hurst exponent for persistence/anti-persistence.
- Mean-reversion strength: estimate AR(1) or OU on de-meaned price; half-life `HL = ln(2)/|ln(φ)|` where `x_t = φ x_{t-1} + ε_t`.
- Momentum/reversal diagnostics: rolling z-scores, moving-average crossovers, short-horizon reversal tests.

### 3) Cross-Asset Structure (Correlations & PCA)
- Build a panel of relevant cross-asset returns (rates, credit, FX, commodities, other indices).
- Compute static and rolling correlations; identify clusters and time-varying co-movements.
- Principal Components: eigen-spectrum to quantify common risk drivers; track loadings of the index on top components.

### 4) Macro Sensitivities (Betas)
- Align macro factors to market dates (e.g., policy rates, curve slope, inflation surprises, growth proxies, dollar, commodities, credit spreads).
- Run OLS with HAC/Newey–West errors: `r_index = α + β_f r_f + ε`, univariate and multivariate.
- Rolling windows (e.g., 252/504) for time-varying betas and `R^2`; stability diagnostics and multicollinearity checks (VIF, regularized/Lasso variants if needed).
- Deliverables: ranked sensitivities with confidence intervals and beta heatmaps over time.

### 5) Regime Detection
- Change-point detection on mean/variance of returns or realized vol (e.g., PELT/BIC) to segment structural shifts.
- Hidden Markov Models (2–3 states) on returns or vol to infer bull/bear/high-vol regimes; report state means/vols and transition matrix.
- Map regimes to macro context and stress episodes; compute regime-conditional performance.

### 6) Seasonality & Calendar Effects
- Day-of-week and month-of-year effects; turn-of-month; holiday and options-expiry proximity.
- Use HAC-robust tests and multiple-testing control; prefer reporting effect sizes with uncertainty rather than binary significance.

### 7) Liquidity & Flow Proxies
- If volume available: turnover and Amihud illiquidity `ILLIQ_t = |r_t|/Volume_t`; otherwise, gap risk (open-to-close vs. close-to-close) as a proxy.
- Track liquidity regimes and relation to realized volatility and drawdowns.

### 8) Tail Risk & Drawdowns
- Rolling max drawdown, recovery time, and pain ratios.
- Extreme Value Theory (POT): tail index and conditional ES for left tail; rolling historical VaR/ES for monitoring.

### 9) Behavioural Signals
- Cross-horizon momentum (3/6/12m) vs. short-term reversal; breadth measures (if constituents available).
- Risk appetite proxies (e.g., credit vs. equity spread changes) to contextualize flows and sentiment.

### 10) Validation & Robustness
- Rolling/out-of-sample checks for all diagnostics; block bootstrap for confidence intervals.
- Sensitivity to lookback windows, sampling frequency, and outlier handling; multiple-testing awareness.

### Outputs
- Tables: macro betas with CIs and `R^2`, correlation/cluster summaries, regime labels and transition stats, seasonality effects.
- Plots: distribution/QQ, ACF/PACF, rolling vol, rolling betas, PCA spectrum/loadings, regime timeline, drawdown and tail metrics.
- Monitoring: current regime, current beta profile, current tail/illiquidity indicators, and an executive summary of risks/opportunities.


```python
if 'nexus_index_bundle' not in globals():
    raise RuntimeError("Run the preprocessing stages to create 'nexus_index_bundle' before executing the market dynamics analysis.")

bundle = nexus_index_bundle
aligned_panel = bundle['aligned_panel']
returns = bundle['stationary']['returns']
diffs = bundle['stationary']['diff']

macro_indicator_keywords = [
    'gdp', 'pmi', 'cpi', 'ppi', 'pce', 'yoy', 'qoq', 'mom', 'monthly', 'net', 'nfp',
    'unemployment', 'sentiment', 'confidence', 'surprise', 'm2', 'industrial', 'manufacturing',
    'housing', 'usurtot', 'nhspatot', 'cesi', 'rstax', 'dgnochng', 'authorized',
    'financial', 'conditions', 'goods', 'production', 'retail'
]
asset_rules = [
    ('Crypto', ['crypto', 'bgci']),
    ('Volatility', ['vix', 'volatility', 'v2x', 'fvsu5c', 'move']),
    ('FX', ['dxy', 'dollar', 'curncy']),
    ('Commodities', ['cf1', ' z', 'commodity', 'metals']),
    ('Credit', ['oas', 'iboxhy', 'iboxig', 'ibox', 'credit', 'leveraged', 'spbdll']),
    ('Rates', ['yield', 'rate', 'usgg10yr', 'usgg', 'shibor', 'estr', 'fdtr', 'bdif5y', 'btsi', 'bcit1t', 'breakeven', 'inflation-li', 'usggbe10']),
    ('Alternative', ['qai', 'multi', 'balanced']),
    ('Equity', ['equity', 'spx', 'sp500', 's&p', 'nikkei', 'nky', 'csi', 'csi500', 'ifbz5', 'shsz300', 'sse50', 'dax', 'djbgicut', 'fnartr', 'es1', 'esz5c', 'esz5p', 'esz', 'emini', 'rpg', 'vfiax', 'vbi', 'vvi', 'vigax', 'gx1', 'eem', 'reit'])
]
macro_keywords = {
    'Growth': ['gdp', 'industrial', 'nfp', 'unemployment', 'pmi', 'manufacturing', 'confidence', 'sentiment'],
    'Inflation': ['cpi', 'inflation', 'ppi', 'pce', 'deflator'],
    'Liquidity': ['m2', 'credit', 'financing'],
    'Policy Rates': ['rate', 'fdtr', 'target', 'estr', 'shibor'],
    'Surprise': ['surprise', 'cesi'],
}

def tokenize_words(name: str) -> set[str]:
    cleaned = ''.join(ch.lower() if ch.isalnum() else ' ' for ch in name)
    return {token for token in cleaned.split() if token}

def classify_asset(name: str) -> str:
    lname = name.lower()
    tokens = tokenize_words(name)
    if any(keyword in lname for keyword in macro_indicator_keywords):
        return 'Macro Indicator'
    for label, keywords in asset_rules:
        for keyword in keywords:
            if ' ' in keyword:
                if keyword in lname:
                    return label
            else:
                if keyword in tokens:
                    return label
    return 'Other Asset'

def classify_macro_bucket(name: str) -> str:
    lname = name.lower()
    for bucket, keywords in macro_keywords.items():
        if any(keyword in lname for keyword in keywords):
            return bucket
    return 'Other Macro'

series_metadata_rows = []
for series_name in aligned_panel.columns:
    series_metadata_rows.append({
        'series': series_name,
        'asset_class': classify_asset(series_name),
        'macro_bucket': classify_macro_bucket(series_name),
    })
series_metadata = pd.DataFrame(series_metadata_rows).set_index('series')

returns_monthly = returns.resample('ME').sum(min_count=1)
diffs_monthly = diffs.resample('ME').mean()

asset_group_returns = {}
for asset_class in sorted(series_metadata['asset_class'].unique()):
    if asset_class in {'Other Asset', 'Macro Indicator'}:
        continue
    cols = series_metadata.index[series_metadata['asset_class'] == asset_class]
    if len(cols) == 0:
        continue
    asset_group_returns[asset_class] = returns_monthly[cols].mean(axis=1)
asset_group_df = pd.DataFrame(asset_group_returns).dropna(how='all')

macro_group_changes = {}
for macro_bucket in sorted(series_metadata['macro_bucket'].unique()):
    if macro_bucket == 'Other Macro':
        continue
    cols = series_metadata.index[series_metadata['macro_bucket'] == macro_bucket]
    shared_cols = [col for col in cols if col in diffs_monthly.columns]
    if not shared_cols:
        continue
    agg = diffs_monthly[shared_cols].mean(axis=1)
    macro_group_changes[macro_bucket] = zscore_normalize(agg)
macro_factors_df = pd.DataFrame(macro_group_changes).dropna(how='all')
macro_factors_df = macro_factors_df.replace([np.inf, -np.inf], np.nan).dropna(how='all')

macro_beta_rows = []
macro_corr_rows = []
for asset_class, series in asset_group_df.items():
    aligned = pd.concat([series.rename('target'), macro_factors_df], axis=1, join='inner').dropna()
    if aligned.empty:
        continue
    X = aligned.iloc[:, 1:].to_numpy()
    y = aligned['target'].to_numpy()
    X = np.column_stack([np.ones(len(X)), X])
    beta, *_ = np.linalg.lstsq(X, y, rcond=None)
    fitted = X @ beta
    resid = y - fitted
    denom = ((y - y.mean()) ** 2).sum()
    r_squared = 1.0 - (resid ** 2).sum() / denom if denom != 0 else np.nan
    for factor, coefficient in zip(aligned.columns[1:], beta[1:]):
        macro_beta_rows.append({
            'asset_class': asset_class,
            'macro_factor': factor,
            'beta': coefficient,
            'r_squared': r_squared,
        })
        corr_val = aligned['target'].corr(aligned[factor])
        macro_corr_rows.append({
            'asset_class': asset_class,
            'macro_factor': factor,
            'correlation': corr_val,
        })
macro_beta_table = pd.DataFrame(macro_beta_rows)
macro_corr_table = pd.DataFrame(macro_corr_rows)

market_dynamics = {
    'series_metadata': series_metadata,
    'asset_group_returns_monthly': asset_group_df,
    'macro_factors': macro_factors_df,
    'macro_beta_table': macro_beta_table,
    'macro_corr_table': macro_corr_table,
}

macro_beta_table.sort_values(['asset_class', 'beta'], ascending=[True, False]).head(10)

```

## 3.1 Macro Impact Interpretation


This block quantifies how each asset bucket co-moves with high-level macro composites using a simple, static multivariate regression and a companion correlation table.

- Inputs
  - `asset_group_returns_monthly`: equal‑weight monthly returns by asset class (averaged across constituent series).
  - `macro_factors`: macro composites built by taking monthly changes of each macro series, averaging within `macro_bucket`, and z‑scoring the aggregate.

- Model
  - For each asset class, run OLS via least squares on the full sample: `y_t = α + Σ_k β_k f_{k,t} + ε_t`, with an intercept.
  - Compute fitted values/residuals and a plain `R^2 = 1 − SSE/TSS` as an explanatory share.
  - Also compute pairwise Pearson correlations between the target and each factor as a directionality sanity check.

- Interpretation
  - Sign: `β_k > 0` implies the asset bucket tends to rise when the macro factor increases; `β_k < 0` implies headwind.
  - Magnitude: because factors are z‑scored, `β_k` is the average return change associated with a +1σ move in factor `k` (in monthly return units).
  - `R^2`: fraction of variation explained by the factor set for that asset bucket; higher values indicate stronger macro linkage.
  - Compare `macro_beta_table` and `macro_corr_table` — they should agree on the sign; large sign mismatches suggest instability or collinearity.

- Robustness & Caveats
  - Static, full‑sample betas can drift across regimes; consider rolling windows to assess stability before acting on signals.
  - Predictors may be collinear across macro buckets; prefer interpretation by groups/trends rather than single coefficients.
  - Levels vs. changes: factors use monthly changes and z‑scores; if the economic mechanism is level‑sensitive (e.g., policy rate level), re‑estimate on levels as a sensitivity.
  - No HAC/Newey–West standard errors are reported here; treat significance qualitatively or extend with robust errors if needed.

- Outputs
  - `macro_beta_table`: columns `[asset_class, macro_factor, beta, r_squared]`.
  - `macro_corr_table`: columns `[asset_class, macro_factor, correlation]`.
  - Use the top positive/negative betas and higher `R^2` groups to summarise macro exposure by asset class.


```python
asset_group_df = market_dynamics['asset_group_returns_monthly']
if asset_group_df.empty:
    raise ValueError('No asset group returns available for cross-asset analysis.')

lags = range(1, 6)
lead_lag_records = []
for src in asset_group_df.columns:
    for dst in asset_group_df.columns:
        if src == dst:
            continue
        base = pd.concat([asset_group_df[src], asset_group_df[dst]], axis=1, join='inner').dropna()
        if base.empty:
            continue
        for lag in lags:
            corr_val = base[src].shift(-lag).corr(base[dst])
            if pd.notna(corr_val):
                lead_lag_records.append({
                    'leader': src,
                    'follower': dst,
                    'lag_months': lag,
                    'lead_correlation': corr_val,
                })
lead_lag_df = pd.DataFrame(lead_lag_records)

if not lead_lag_df.empty:
    top_leads = lead_lag_df.sort_values('lead_correlation', ascending=False).head(10)
else:
    top_leads = pd.DataFrame()

market_dynamics['lead_lag_table'] = lead_lag_df
market_dynamics['top_leads'] = top_leads

top_leads
```

## 3.2 Cross-Asset Lead/Lag Insights


We measure whether moves in one asset bucket tend to precede moves in another by correlating leading returns with contemporaneous follower returns.

- Method
  - For each ordered pair `(leader, follower)` and lags `L ∈ {1,…,5}` months, compute `corr(r^{leader}_{t+L}, r^{follower}_t)` on the overlapping sample.
  - Record the correlation and surface the top relationships by `lead_correlation`.

- Interpretation
  - Positive correlation at lag `L` suggests the leader tends to move `L` months before the follower in the same direction.
  - Use stable, repeatable chains (e.g., rates → equities, USD → commodities) across several lags/subperiods as tactical early‑warning signals.

- Robustness & Caveats
  - Correlation ≠ causation; validate with sub‑sample checks and economic rationale.
  - Outliers and overlapping windows can inflate correlations; winsorization or Spearman correlations can help.
  - For formal inference, add Granger causality tests on the same panel as a follow‑up.

- Outputs
  - `lead_lag_table`: `[leader, follower, lag_months, lead_correlation]`.
  - `top_leads`: top entries sorted by `lead_correlation`.


```python
regimes = nexus_index_bundle['regimes']
returns = nexus_index_bundle['stationary']['returns']
series_metadata = market_dynamics['series_metadata']

regime_distribution = regimes.apply(lambda col: col.value_counts(normalize=True)).T.fillna(0.0)

regime_performance_rows = []
for series_name in regimes.columns:
    if series_name not in returns.columns:
        continue
    joined = pd.DataFrame({
        'return': returns[series_name],
        'regime': regimes[series_name],
    }).dropna()
    if joined.empty:
        continue
    grouped = joined.groupby('regime')['return'].mean()
    for regime_label, avg_ret in grouped.items():
        regime_performance_rows.append({
            'series': series_name,
            'asset_class': series_metadata.loc[series_name, 'asset_class'],
            'regime': regime_label,
            'avg_return': avg_ret,
        })
regime_performance = pd.DataFrame(regime_performance_rows)

market_dynamics['regime_distribution'] = regime_distribution
market_dynamics['regime_performance'] = regime_performance

regime_distribution.head()
```

## 3.3 Regime Diagnostics Reading


We classify regimes using a simple moving‑average crossover and then summarize both the time spent in each state and the typical payoff per state.

- Regime labelling
  - `detect_market_regimes(series, fast=20, slow=100)`: bull if `MA_fast > MA_slow`, bear if `MA_fast < MA_slow`, else stagnant.
  - Applied per series; parameters target medium‑term trend vs. noise.

- Diagnostics
  - `regime_distribution`: share of observations per regime and series (time in state).
  - `regime_performance`: average arithmetic return conditioned on the inferred regime.

- Interpretation
  - High bull share with positive bull returns supports trend‑following overlays; negative bear returns favour hedges or gross‑down in risk.
  - Large stagnant share suggests range‑bound dynamics: mean‑reversion and carry profiles dominate.

- Caveats & Extensions
  - MA crossovers ignore volatility regime and jump risk; consider HMM or change‑point models for richer state dynamics.
  - Validate parameters with sensitivity tests (fast/slow windows) and check alignment with known market episodes.


```python
features = nexus_index_bundle['features']
anomalies = nexus_index_bundle['anomalies']
returns = nexus_index_bundle['stationary']['returns']

behaviour_rows = []
if isinstance(features.columns, pd.MultiIndex) and 'rsi14' in features.columns.get_level_values(1):
    rsi_panel = features.xs('rsi14', level=1, axis=1)
    next_returns = returns.shift(-5)
    for series_name in rsi_panel.columns:
        rsi_series = rsi_panel[series_name].dropna()
        if rsi_series.empty:
            continue
        overbought = rsi_series[rsi_series > 70].index
        oversold = rsi_series[rsi_series < 30].index
        overbought_ret = next_returns.loc[overbought, series_name].mean()
        oversold_ret = next_returns.loc[oversold, series_name].mean()
        behaviour_rows.append({
            'series': series_name,
            'signal': 'RSI>70 forward 5d',
            'avg_forward_return': overbought_ret,
            'signal_count': len(overbought),
        })
        behaviour_rows.append({
            'series': series_name,
            'signal': 'RSI<30 forward 5d',
            'avg_forward_return': oversold_ret,
            'signal_count': len(oversold),
        })

if anomalies is not None:
    for series_name in anomalies.columns:
        flagged = anomalies[series_name] == 1
        if flagged.any() and series_name in returns.columns:
            subsequent = returns[series_name].shift(-1)
            behaviour_rows.append({
                'series': series_name,
                'signal': 'Anomaly flag next day',
                'avg_forward_return': subsequent[flagged].mean(),
                'signal_count': int(flagged.sum()),
            })

behaviour_table = pd.DataFrame(behaviour_rows)
market_dynamics['behaviour_table'] = behaviour_table

behaviour_table.head(10)
```

### 4.4.3 Asset-Class Regime Timelines (actual start dates)
For each asset class, we construct an equal‑weight price composite from constituent series (each normalized to its own first valid observation), infer regimes via MA(20/100) on that composite, and plot the timeline from the asset class’s true first available date (not truncated by later cleaning windows). Colors: bull = green, bear = red, stagnant = gray.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Guard: require bundle and metadata
if 'nexus_index_bundle' not in globals():
    raise RuntimeError('nexus_index_bundle not found. Run the preprocessing pipeline (2.x) first.')

bundle = nexus_index_bundle
aligned_panel = bundle.get('aligned_panel', pd.DataFrame())
series_metadata = market_dynamics.get('series_metadata', pd.DataFrame()) if 'market_dynamics' in globals() else pd.DataFrame()
if aligned_panel.empty or series_metadata.empty:
    raise RuntimeError('Aligned panel or series metadata unavailable. Ensure earlier sections executed successfully.')

# Helper: normalize each series to 100 at its own first valid observation
def normalize_to_first_valid(s: pd.Series) -> pd.Series:
    s = s.copy()
    # Use first non-NaN and non-zero value as base
    valid = s.replace(0.0, np.nan).dropna()
    if valid.empty:
        return s * np.nan
    base_idx = valid.index[0]
    base_val = valid.iloc[0]
    out = (s / base_val) * 100.0
    out.loc[:base_idx] = out.loc[:base_idx]  # preserve NaNs before base
    return out

asset_classes = [c for c in sorted(series_metadata['asset_class'].dropna().unique())]
if not asset_classes:
    raise RuntimeError('No asset classes found in metadata.')

# Build composite per asset class
composites: dict[str, pd.Series] = {}
for ac in asset_classes:
    names = series_metadata.index[series_metadata['asset_class'] == ac]
    names = [n for n in names if n in aligned_panel.columns]
    if not names:
        continue
    norm_df = pd.concat([normalize_to_first_valid(aligned_panel[n]) for n in names], axis=1)
    norm_df.columns = names
    composite = norm_df.mean(axis=1, skipna=True)
    composites[ac] = composite

if not composites:
    raise RuntimeError('Could not build any asset-class composites from aligned_panel and metadata.')

# Infer regimes on composites using existing helper
regime_colors = {'bull': '#2ca02c', 'bear': '#d62728', 'stagnant': '#7f7f7f'}
n = len(composites)
ncols = 2
nrows = int(np.ceil(n / ncols))
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(12, max(3, 2.2 * nrows)), squeeze=False)
axes = axes.flatten()
for idx, (ac, comp) in enumerate(sorted(composites.items())):
    ax = axes[idx]
    # Start from the asset class's actual first available date
    start = comp.first_valid_index()
    if start is None:
        ax.set_visible(False)
        continue
    comp_slice = comp.loc[start:]
    # Regime from price composite (MA 20/100, min_periods=1 in helper)
    reg = detect_market_regimes(comp_slice)
    # Background shading by contiguous regime segments
    seg_id = (reg != reg.shift()).cumsum()
    for seg, idxs in reg.groupby(seg_id).groups.items():
        seg_lab = reg.loc[idxs[0]] if isinstance(idxs, list) else reg.loc[idxs].iloc[0]
        lab = str(seg_lab) if pd.notna(seg_lab) else 'stagnant'
        col = regime_colors.get(lab, '#7f7f7f')
        ixs = reg.index[reg.index.isin(idxs)] if isinstance(idxs, list) else reg.loc[idxs].index
        if len(ixs) == 0:
            continue
        ax.axvspan(ixs[0], ixs[-1], color=col, alpha=0.15, linewidth=0)
    # Price line
    ax.plot(comp_slice.index, comp_slice.values, color='black', linewidth=1.0)
    ax.set_title(f'{ac} — Regime timeline')
    ax.set_ylabel('Index (norm=100)')
    ax.grid(True, alpha=0.2, linestyle='--')

# Hide any leftover axes
for j in range(idx + 1, len(axes)):
    axes[j].set_visible(False)

# Legend
handles = [Patch(facecolor=regime_colors[k], edgecolor='none', alpha=0.3, label=k.title()) for k in ['bull', 'bear', 'stagnant']]
fig.legend(handles=handles, loc='upper center', ncols=3, frameon=False)
fig.tight_layout(rect=(0, 0, 1, 0.94))
plt.show()
```

#### How asset‑class regimes are measured with multiple instruments

- Composite‑based regime (used above)
  - Normalize each constituent series to 100 at its own first valid observation, then average (equal‑weight) into a daily composite.
  - Start the class timeline at the composite’s first valid date, so early constituents are not truncated by later series.
  - Classify regimes on the composite with a simple MA(20/100) rule: bull if fast>slow, bear if fast<slow, else stagnant.
  - Rationale: reduces idiosyncratic noise, handles staggered launches, and gives a single interpretable path.

- Alternatives (useful cross‑checks)
  - Majority vote: compute per‑instrument regimes with the same rule; assign the asset‑class regime as the daily majority (optionally require ≥60% agreement; otherwise mark "mixed/stagnant").
  - Return composite: average returns (equal/vol/liquidity‑weighted), cumulate to a pseudo‑price, then classify with MA(20/100).
  - Multivariate: use PC1 of the class return panel or a multivariate HMM to infer hidden states and transitions.

- Practical choices
  - Weighting: swap equal‑weight for inverse‑vol, liquidity, or cap weights if appropriate.
  - Data gaps: require a minimum number of active instruments per day; winsorize or smooth to reduce spurious flips.
  - Sensitivity: MA windows control responsiveness; validate across sub‑samples and horizons.

In this notebook, the composite‑based approach is the default for clarity. Consider adding a majority‑vote series as a robustness overlay if class membership or liquidity changes materially over time.
